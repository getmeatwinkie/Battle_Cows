```python
import random
import time

def displayIntro():
    print("You wake up, throbbing headache, and a sore body,")
    print("You are standing on all four legs,")
    print("You are a Cow.")
    print("But you are not just any regular old cow,")
    print("You...are a BATTLE COW")
    print()
    
def chooseBreed():
    breed = ''
    while breed != 'angus' and breed != 'holstein':
        print("Which breed are you? (angus or holstein)")
        print("Angus-  +6 strength, +4 speed, +5 beef")
        print("Holstein-  +4 strength, +6 speed, +5 dairy")
        breed = input()
        
    return breed

def checkBreed(chosenBreed):
    print("Ah yes! A " +chosenBreed+ " Battle Cow!")
    if chosenBreed == 'angus':
        print("As an Angus Battle Cow, you are quite beefy, what would you like to do first? ")
    else:
        print("As a Holstein Battle Cow, you are quite milky, what would you like to do first? ")
        
def firstchoice():
    choice1 = ''
    while choice1 != 'die' and choice1 != 'walk' and choice1 != 'implode' and choice1 != 'run' and choice1 != 'jump' and choice1 != 'jump forward' and choice1 != 'cast a spell' and choice1 != 'become human' and choice1 != 'use the restroom' and choice1 != 'float' and choice1 != 'start a religion' and choice1 != 'get a job':
        choice1 = input()
        
    return choice1

def checkChoice():
    choice1 = firstchoice()
    if choice1 == 'die':
        print('You chose to die? How do you even do that? Whatever. GAME OVER')
    elif choice1 == 'walk':
        print('You walk for 5 minutes, and arrive upon an Angry Farmhand. I think he wants you to put the hurt on him')
    elif choice1 == 'run':
        print('You run for 5 minutes, and arrive upon an Angry Farmhand. OOOH a fight scene!')
    elif choice1 == 'jump':
        print('You attempt to jump upwards, you fail because you are a cow and they cant do that')
    elif choice1 == 'jump forward':
        print('You jump forward like 4 feet. That did pretty much nothing')
    elif choice1 == 'cast a spell':
        print('You just casted a spell! It turned a bush you were looking at into an apple pie. It was tasty.')
    elif choice1 == 'become human':
        print('You somehow turned yourself back into a human. I guess you win...Good job.')
    elif choice1 == 'use the restroom': 
        print('You just did that...please dont do it again thats gross.')
    elif choice1 == 'float': 
        print('You cant do that youre a c- WHAT HOW ARE YOU FLOATING?!')
    elif choice1 == 'start a religion': 
        print('You have started a religion focused on you. You are your own God. YOU WIN')
    elif choice1 == 'get a job': 
        print('You applied for a job, however, you are a cow and are immediately turned away when you show up for the interview.')
    elif choice1 == 'implode': 
        print('There is nothing left but chunks of cow and emptyness.')

        
playAgain = 'yes'
while playAgain == 'yes' or playAgain == 'y':
    
    displayIntro()
    
    breed_choice = chooseBreed()
    
    checkBreed(breed_choice)
    
    checkChoice()
    
    print('Do you want to play again? (yes or no)')
    playAgain = input()
```
